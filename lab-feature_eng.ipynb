```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check the shape of your data**

**Check for missing values**

```python
#your code here

spaceship.isna().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
#your code here

spaceship = spaceship.dropna()
```

```python
spaceship.isna().sum()
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
#your code here

spaceship["Cabin"] = spaceship["Cabin"].str.split("/").str[0]

spaceship["Cabin"].head(10)
```

- Drop PassengerId and Name

```python
#your code here

spaceship = spaceship.drop(columns=["PassengerId", "Name"])

spaceship.head()
```

- For non-numerical columns, do dummies.

```python
#your code here

columnas_categoricas = spaceship.select_dtypes(include="object").columns
spaceship = pd.get_dummies(spaceship, columns=columnas_categoricas, drop_first=True)

spaceship.head()
```

**Perform Train Test Split**

#### Utilizamos el train_test_split sin "stratify=y" 

```python
#your code here

target = spaceship['Transported']
features = spaceship.drop(columns=["Transported"])

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=42)
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
#your code here

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

- Evaluate your model's performance. Comment it

```python
#your code here

y_pred = knn.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

#### Utilizamos el train_test_split con "stratify=y" 

```python
target = spaceship['Transported']
features = spaceship.drop(columns=["Transported"])

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=42, stratify=target)

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
